Kmeans

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime1=pd.read_csv("crime_data.csv")
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

```python
crime1
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

DBSCAN


```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime3=pd.read_csv("crime_data.csv")
```

```python
crime3.info()
```

```python
df=crime3.iloc[:,1:5]
df
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
#noisy samples are given the label -1
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
clustered=pd.concat([crime3,cl],axis=1)
```

```python
clustered
```
